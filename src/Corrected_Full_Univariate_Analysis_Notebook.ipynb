```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
sns.set(style="whitegrid")
    
```

```python


# Load the dataset
df = pd.read_excel('../assets/consumo_material_clean.xlsx')
df.head()
    
```

```python

# Count plot for the categorical column: CODIGO
plt.figure(figsize=(10, 6))
sns.countplot(x=df['CODIGO'].dropna())
plt.title('Count of Categories in CODIGO')
plt.xlabel('CODIGO')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
                
```

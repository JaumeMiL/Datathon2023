```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set style for plots
sns.set(style="whitegrid")
    
```

```python

# Load the dataset
df = pd.read_excel('../assets/consumo_material_clean.xlsx')

# Convert 'FECHAPEDIDO' to datetime format and handle potential inconsistencies
df['FECHAPEDIDO'] = pd.to_datetime(df['FECHAPEDIDO'], errors='coerce', format='%d/%m/%y')

# Drop rows where 'FECHAPEDIDO' couldn't be converted
df.dropna(subset=['FECHAPEDIDO'], inplace=True)

# Extract 'Year' from 'FECHAPEDIDO'
df['Year'] = df['FECHAPEDIDO'].dt.year
    
```

```python

# Bar chart of the sum of CANTIDADCOMPRA per year
sum_per_year = df.groupby('Year')['CANTIDADCOMPRA'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='CANTIDADCOMPRA', data=sum_per_year)
plt.title('Total CANTIDADCOMPRA per Year')
plt.xlabel('Year')
plt.ylabel('Total CANTIDADCOMPRA')
plt.xticks(rotation=45)
plt.show()
                
```

```python

# Bar chart of the sum of UNIDADESCONSUMOCONTENIDAS per year
sum_per_year = df.groupby('Year')['UNIDADESCONSUMOCONTENIDAS'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='UNIDADESCONSUMOCONTENIDAS', data=sum_per_year)
plt.title('Total UNIDADESCONSUMOCONTENIDAS per Year')
plt.xlabel('Year')
plt.ylabel('Total UNIDADESCONSUMOCONTENIDAS')
plt.xticks(rotation=45)
plt.show()
                
```

```python

# Bar chart of the sum of PRECIO per year
sum_per_year = df.groupby('Year')['PRECIO'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='PRECIO', data=sum_per_year)
plt.title('Total PRECIO per Year')
plt.xlabel('Year')
plt.ylabel('Total PRECIO')
plt.xticks(rotation=45)
plt.show()
                
```

```python
# Count of rows per year
rows_per_year = df.groupby('Year').size().reset_index(name='Count')

# Bar chart of the number of rows per year
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='Count', data=rows_per_year)
plt.title('Number of Rows per Year')
plt.xlabel('Year')
plt.ylabel('Number of Rows')
plt.xticks(rotation=45)
plt.show()
```

```python

# Bar chart of the sum of IMPORTELINEA per year
sum_per_year = df.groupby('Year')['IMPORTELINEA'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='IMPORTELINEA', data=sum_per_year)
plt.title('Total IMPORTELINEA per Year')
plt.xlabel('Year')
plt.ylabel('Total IMPORTELINEA')
plt.xticks(rotation=45)
plt.show()
                
```

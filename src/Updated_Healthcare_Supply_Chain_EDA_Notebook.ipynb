```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set style for plots
sns.set(style="whitegrid")
    
```

```python

# Load the dataset
df_train = pd.read_csv('../assets/train.csv')

# Create a 'Date' column by combining 'AÑO' and 'MES'
df_train['Date'] = pd.to_datetime(df_train['AÑO'].astype(str) + '-' + df_train['MES'].astype(str))

# Extract year from 'Date'
df_train['Year'] = df_train['Date'].dt.year
    
```

```python

# Yearly trend in 'CANTIDADCOMPRA'
yearly_cantidad_compra = df_train.groupby('Year')['CANTIDADCOMPRA'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='CANTIDADCOMPRA', data=yearly_cantidad_compra)
plt.title('Yearly Trend in CANTIDADCOMPRA')
plt.xlabel('Year')
plt.ylabel('Total CANTIDADCOMPRA')
plt.xticks(rotation=45)
plt.show()
    
```

```python

# Yearly trend in 'IMPORTELINEA'
yearly_importe_linea = df_train.groupby('Year')['IMPORTELINEA'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='IMPORTELINEA', data=yearly_importe_linea)
plt.title('Yearly Trend in IMPORTELINEA')
plt.xlabel('Year')
plt.ylabel('Total IMPORTELINEA')
plt.xticks(rotation=45)
plt.show()
    
```
